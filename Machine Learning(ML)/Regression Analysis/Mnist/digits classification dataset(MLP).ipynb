# Visualize datasets

```python
import matplotlib.pyplot as plt
def plot_image(image):
    fig = plt.gcf()
    fig.set_size_inches(2, 2)
    plt.imshow(image, cmap = 'binary')
    plt.show()

plot_image(x_train_image[0])
y_train_label[0]
```
